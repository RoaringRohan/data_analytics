```python
import pandas as pd
```

```python
df = pd.read_csv('../1_Data/extracted_dataset.csv')

df.head()
```

```python
cols = ['weather_type', 'hour', 'month']

df.drop(cols, inplace = True, axis = 1)

df.head()
```

```python
# Assuming df is your DataFrame and 'time_column' is the name of your time column
df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Fill any NaT values if necessary (optional, depending on your data quality requirements)
df['Time'] = df['Time'].fillna(pd.to_datetime(df['Time'], errors='coerce'))

df.head()
```

```python
df.dtypes
```

```python
# Splitting multiple values and creating separate rows
df_time = df[['Time']].copy()

df_time.head()
```

```python
df_time.to_csv('../1_Data/transformed_time_dataset.csv', index=False)
```

```python
df_weather_measurement = df.drop(columns=['Time'])

df_weather_measurement.head()
```

```python
df_weather_measurement.to_csv('../1_Data/transformed_weather_measurement_dataset.csv', index=False)
```
