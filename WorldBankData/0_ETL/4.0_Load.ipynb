```python
#Importing pandas

import pandas as pd
```

```python
#Read dataform
df = pd.read_csv('../1_Data/transformed_dataset.csv')

df.head()      
```

```python
selected_columns = ['Country Name', 'Country Code', 'Region Code']
countries = []

for i in range(len(df)):
    selected_data = df.loc[i, selected_columns]
    countries.append(tuple(selected_data))

countries
```

```python
file = open('4.1_countries.txt', 'w')

for row in countries:
    file.write(str(row) + ',' + '\n')

file.close()
```

```python
population_selected_columns = ['Country Name', ' Population, total ', 'Population CGR 1960-2015']
populations = []

for i in range (len(df)):
    population_selected_data = df.loc[i, population_selected_columns]
    populations.append(tuple(population_selected_data))

populations
```

```python
file = open('4.2_populations.txt', 'w')

for row in populations:
    file.write(str(row) + ',' + '\n')

file.close()
```
