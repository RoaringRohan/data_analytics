```python
import pandas as pd
```

```python
#Read dataform created after extracting process finished
df = pd.read_csv('../1_Data/extracted_dataset.csv')

df.head()
```

```python
#USE ONLY WHEN YOU WANT TO REMOVE COLUMNS

#cols = ['Region Code', 'Country Code']

#df.drop(cols, inplace = True, axis = 1)

#df.head()
```

```python
#Removing commas from columns
df['GDP, PPP (current international $)'] = df['GDP, PPP (current international $)'].str.replace(',', '')
df[' Population, total '] = df[' Population, total '].str.replace(',', '')

df.head()
```

```python
#Removing percent signs from columns
df['Population CGR 1960-2015'] = df['Population CGR 1960-2015'].str.replace('%', '')
df['Population CGR 1960-2015'] = pd.to_numeric(df['Population CGR 1960-2015']) / 100

df['Popltn Largest City % of Urban Pop'] = df['Popltn Largest City % of Urban Pop'].str.replace('%', '')
df['Popltn Largest City % of Urban Pop'] = pd.to_numeric(df['Popltn Largest City % of Urban Pop']) / 100

df.head()
```

```python
#If you need to adjust the variable types before preparing for loading, do it here
df = df.astype({'GDP, PPP (current international $)':'float'})
df = df.astype({' Population, total ':'float'})
df = df.astype({'Population CGR 1960-2015':'float'})
df = df.astype({'Popltn Largest City % of Urban Pop':'float'})
```

```python
#Save df as csv for next file
df.to_csv('../1_Data/transformed_dataset.csv', index=False)
```
