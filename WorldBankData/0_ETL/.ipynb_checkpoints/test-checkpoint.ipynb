```python
import pandas as pd
```

```python
df = pd.read_csv('2015 World Bank Data.csv')

df.head()
```

```python
#USE ONLY WHEN YOU WANT TO REMOVE COLUMNS

cols = ['Region Code', 'Country Code']

df.drop(cols, inplace = True, axis = 1)

df.head()
```

```python
df.isnull().sum()
```

```python
#USE ONLY TO REMOVE TUPLES IN ROWS WITH NULL DATA FROM SPECIFIC COLUMNS

df.dropna(subset=['Population CGR 1960-2015', 'Literacy rate, adult female (% of females ages 15 and above)', 'Exports of goods and services (% of GDP)', 'Popltn Largest City % of Urban Pop'], inplace = True)

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.dtypes
```
