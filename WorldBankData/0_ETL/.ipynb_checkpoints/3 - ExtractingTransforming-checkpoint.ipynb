```python
import pandas as pd
```

```python
df = pd.read_csv('2015 World Bank Data.csv')

df.head()
```

```python
cols = ['Region Code', 'Country Code', 'GDP, PPP (current international $)']

df.drop(cols, inplace = True, axis = 1)

df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Population CGR 1960-2015', 'Literacy rate, adult female (% of females ages 15 and above)', 'Exports of goods and services (% of GDP)', 'Popltn Largest City % of Urban Pop'], inplace = True)

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.dtypes
```

```python
y = []

for i in range (len(df)):
    x = tuple(df.iloc[i])
    y.append(x)

y
```

```python
file = open('WorldBankData.txt', 'w')

for tuple in y:
    file.write(str(tuple) + ',' + '\n')

file.close()
```
