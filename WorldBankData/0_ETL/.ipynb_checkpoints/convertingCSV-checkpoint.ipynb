```python
import pandas as pd
```

```python
df = pd.read_csv('2015 World Bank Data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
y = []

for i in range (len(df)):
    x = tuple(df.iloc[i])
    y.append(x)

y
```

```python
file = open('WorldBankData.txt', 'w')

for tuple in y:
    file.write(str(tuple) + ',' + '\n')

file.close()
```
