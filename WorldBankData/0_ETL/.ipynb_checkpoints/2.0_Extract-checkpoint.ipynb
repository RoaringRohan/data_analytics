```python
#Importing pandas

import pandas as pd
```

```python
#This line will create a dataform which will read the CSV file from Kaggle

#Read dataform
df = pd.read_csv('../1_Data/raw_dataset.csv')

df.head()             
```

```python
#Checks for how many tuples have null values in them and organizes by column
df.isnull().sum()
```

```python
#USE ONLY TO REMOVE TUPLES IN ROWS WITH NULL DATA FROM SPECIFIC COLUMNS

df.dropna(subset=['Population CGR 1960-2015', 'Literacy rate, adult female (% of females ages 15 and above)', 'Exports of goods and services (% of GDP)', 'Popltn Largest City % of Urban Pop'], inplace = True)

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
#Finds out the python data type for each column
df.dtypes
```

```python
#Save df as csv for next file
df.to_csv('4 - extracted.csv', index=False)
```
